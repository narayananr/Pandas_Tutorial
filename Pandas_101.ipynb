### Python Pandas Introduction

Here is a short introduction to Pandas using gene expression data. The gene expression data is from the publication and downloaded from  ftp://ftp.ncbi.nlm.nih.gov/geo/series/GSE45nnn/GSE45684/suppl/GSE45684_FounderStrainsamples_alignedtoStraingenomes_processedmatrix.txt.gz.
    
Just for conveneience the downloaded file is renamed to a simpler name.

```python
## import pandas
import pandas as pd
```

#### How to read a tsv file into Pandas Dataframe?

```python
exp_df=pd.read_csv("GSE45684_FounderStrainsamples_exp.txt", sep="\t")
exp_df.head()
```

Note that the dataframe has column names and the second row of the data frame is strain name. We can skip the second row while reading the file as dataframe with Pandas. 

#### How to skip a row while readling tsv file into Pandas Dataframe?

```python
exp_df=pd.read_csv("GSE45684_FounderStrainsamples_exp.txt", sep="\t", skiprows=[1])
exp_df.head()
```

#### How to check the shape or size of a data frame?

```python
exp_df.shape
```

Note that the first column of the dataframe is gene names and it is incorrectly named as Sample. Often we may want to remove a column (or more columns). Let us see how to do that.

#### How to remove one  column (or more columns) from Pandas dataframe?

```python
exp_df.drop(['Sample'],axis = 1).head()
```

Insteadof completely deleting the Sample column, let us make that column as rownames of the dataframe.

#### How to assign a column as row names?

```python
exp_df=exp_df.set_index(['Sample'])
exp_df.head()
```

Note that we have successfully used the column as row names and now the dataframe has 128 columns instead of 129. However, the index or rownames still have a name "Sample". 

#### How to remove name of index in Pandas Dataframe?

```python
exp_df.index.name=None
print exp_df.shape
exp_df.head()
```

Now the data frame is simply a matrix of gene expression with sample names as columns and gene ID as row names. Let us explore the data frame.
